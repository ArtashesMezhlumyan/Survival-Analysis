```python
from lifelines import WeibullFitter,\
                      ExponentialFitter,\
                      LogNormalFitter,\
                      LogLogisticFitter,\
                      WeibullAFTFitter,\
                      LogNormalAFTFitter,\
                      LogLogisticAFTFitter
```

```python
df = pd.read_csv('telco.csv')
df.head(5)
```

```python
# indexing by ID
df.set_index('ID', inplace=True)
```

```python
# churn: yes-1, no-0
df['churn'] = pd.Series(np.where(df.churn.values == "Yes", 1, 0), df.index)
```

```python
# encoding the categorical variables into dummies: 1,0
encode_cols = ['region', 'marital', 'ed', 'retire', 'gender', 'voice', 'internet', 'forward', 'custcat']
survival_df = pd.get_dummies(df, columns=encode_cols, prefix=encode_cols, drop_first=True)
#we cannot have zeros we will make thos 0.0000001
survival_df["tenure"] = np.where(survival_df["tenure"] == 0, 0.000001, survival_df["tenure"])
survival_df.head()
```

```python
wf = WeibullFitter()
ef = ExponentialFitter()
lnf = LogNormalFitter()
llf = LogLogisticFitter()
```

```python
fig, ax = plt.subplots(figsize=(16, 8))
# Fitting and plotting
for model in [wf, ef, lnf, llf]:
    model.fit(durations = survival_df["tenure"], event_observed = survival_df["churn"])
    model.plot_survival_function()
    print("The AIC value for", model.__class__.__name__, "is",  model.AIC_)
    print("The BIC value for", model.__class__.__name__, "is",  model.BIC_)
    model.print_summary()
    print('='*100)
```

### Building models using AFT fitters

```python
wf_aft = WeibullAFTFitter()
lnf_aft = LogNormalAFTFitter()
llf_aft = LogLogisticAFTFitter()
```

```python
#same logic for no zeros
survival_df["tenure"] = np.where(survival_df["tenure"] == 0, 0.000001, survival_df["tenure"])
fig, ax = plt.subplots(figsize=(16, 8))


# Fitting and ploting
i = 1
for model in [wf_aft, lnf_aft, llf_aft]:
    model.fit(survival_df, duration_col = "tenure", event_col = "churn")
    if i == 1:
        label = "WeibullAFTFitter"
    elif i == 2:
        label = "LogNormalAFTFitter"
    elif i == 3:
        label = "LogLogisticAFTFitter"
    i = i + 1
    model.print_summary()
    print('='*100)
    plt.plot(model.predict_survival_function(survival_df.loc[1]), label = label)
    plt.legend()
```

#### The best fitter is the one that has the lowest AIC nad BIC. In our case we can observe that LogNormalFitter performs the best as it has the lowest AIC and BIC scores, moreover its log-likelihood value is the highest. Thus, we will continue with LogNormalFitter as our AFT Fitter.

```python
logn_aft = LogNormalAFTFitter()
logn_aft.fit(survival_df, duration_col='tenure', event_col='churn')
logn_aft.print_summary()
```

####  For a feature to significant, it must have a p value of less than 0.05. From the above observations we can claim that address, age, custcat_E-service, custcat_Plus service, custcat_Total service, internet_Yes, marital_Unmarried, voice_Yes are the significant features. At the same time the coefficient's column shows what impact the features have on the survival time of a customer. If the coefficient is positive => increases, if negative=> decreases. So in the survival_df below, I will keep the ones that are significant.


```python
survival_df = survival_df[["tenure", "churn", "address", "age", "custcat_E-service", "custcat_Plus service", "custcat_Total service", "internet_Yes", "marital_Unmarried", "voice_Yes"]]
```

```python
logn_aft = LogNormalAFTFitter()
logn_aft.fit(survival_df, duration_col='tenure', event_col='churn')
logn_aft.print_summary()
```

## CLV

```python
pred_clv = logn_aft.predict_survival_function(survival_df)
pred_clv
```

```python
# keep 1 year
pred = pred_clv.loc[1:12, :]
pred
```

```python
MM = 1300
r = 12
for col in range(1, len(pred.columns)+1):
    for row in range(1, 13):
        pred[col][row] = pred[col][row] / (1 + r / 12)**(row - 1)
df['CLV'] = MM * pred.sum(axis = 0)
df.head()
```

```python
clv = sns.displot(df, x = 'CLV', kind = 'kde', hue = 'custcat', weights = 16, height=8)
clv.set(title = 'Customer Category CLV density')
```

#### From the plot above we can observe that "Plus Service" category customers are the most likely to survive.

```python
# CLV for customers that left the company within a year
average_clv_left = df["CLV"].loc[(df["tenure"] <= 12) & (df["churn"] == 1)].mean()
print(average_clv_left)

# yearly CLV of cutomers that left
clv_yearly = df["CLV"].loc[(df["tenure"] <= 12) & (df["churn"] == 1)].sum()
print(clv_yearly)
```

```python
pred = pred.T.iloc[:, -1].values
```

```python
# I define 0.8 threshold for the survival probability to be considered at risk customer
budget = df["CLV"].loc[(df["tenure"] <= 12) & (pred < 0.8)].sum()
print(budget)
```

Based on the coefficient of our model we can say that one unit increase in covariate indicates that the mean/median survival time will change by a factor of exp(coefficient). If the coefficient is positive, then the exp(coefficient) will be > 1, which will increase the incident/event time. Similarly, a negative coefficient will reduce the mean/median survival time.Therefore a unit increase in age, address results in increase in survival time by 4%. If a person is unmarried (base = married) then survival time will decrease by 36%, and so on.
Being a valuable customer means having a high CLV <=> bringing more revenue/provit over their lifetime to the company. Based on the above CLV density plots within different segments I would say that customers that married/not retired/are subscribers of plus service can be considered valuable (gender does not make much difference).
I defined a threshold of 0.8 for the probability of survival of the customer => if a customer has less than 0.8 prob of survival, then the customer is at-risk. Finally, by summing up the expected values for those customers, we estimate the total annual value of the at-risk subscriber base. This value represents the maximum annual budget that could be spent on retention activities while still generating a positive return on investment.
There can be several ways to address the retention problem. For instance, improving customer service, offering loyalty programs like special promotions or discounts for some at-risk customers.
